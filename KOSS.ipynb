```python
import pandas as pd
import numpy as np
```

#### Loading the Dataset

```python
df=pd.read_csv("/Users/riyasm/Dataset.csv")
df
```

#### Seperating Categorical and Numerical columns

```python
df1=df[['Age','Salary']]
X=df[['Country','Purchased']]
```

### Genaral code for One- hot encoding for any Data Frame from scratch

```python
def one_hot_encoding(X):
    # Creating a temporary single zero columned Data Frame
    x1=pd.DataFrame(np.zeros(len(X)),dtype='int',columns=['Temp_col'])

   # Creating a M*N Data Frame(x1) with all values=0
   # M = no of rows in X & N = sum of all unique categorical values in the given Data Frame
    a=list(X.columns)
    b=len(X)
    c=len(X.columns)
    for i in a:
        x2=pd.DataFrame(np.zeros([len(X),len(X[i].unique())],dtype='int'),columns=X[i].unique()).add_prefix(i+'_')
        x1=pd.concat([x1,x2],axis=1)

    # Updating values in the Data Frame(x1)
    for i in range(b):
        y=list(np.zeros(c,dtype='int'))
        for j in range(c):
            y[j]=X.iloc[i,j]
            x1.loc[i,a[j] +'_'+ y[j]]+=1
    x1.drop("Temp_col",axis=1,inplace=True)
    return x1

x1=one_hot_encoding(X)
```

#### Concatinating both categorical and numerical columns

```python
X_final=pd.concat([x1,df1],axis=1)
X_final
```

```python
X_final.to_csv('Final_dataset.csv', index=True)
```
